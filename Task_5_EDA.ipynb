```python
# Importing the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the seaborn style for charts
sns.set(style="darkgrid")
```

```python
# Loading the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Showing first few rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Value counts of the Survived column
df['Survived'].value_counts()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()
```

```python
# Only select numeric columns
numeric_df = df.select_dtypes(include='number')

# Create the heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna())
plt.show()
```

```python
## My Summary of Findings:

- Most passengers are aged between 20 and 40.
- Some passengers paid very high fares (outliers).
- 'Cabin' and 'Age' have missing values.
- Fare and Pclass have some correlation with Survived.
```
